```python
# Replacing the missing values with mean 
```

```python
# replacing missing values in "normalized-losses" 
avg_nor_loss = file["normalized-losses"].astype("float").mean()
file["normalized-losses"].replace(numpy.NaN, avg_nor_loss, inplace = True)

# Replacing the missing values in "num-of-doors" with the most frequent number of doors
most_freq_doors = file["num-of-doors"].value_counts().idxmax()
file["num-of-doors"].replace(numpy.NaN, most_freq_doors, inplace = True)

# Replacing missing values in "bore"
avg_bore = file["bore"].astype("float").mean()
file["bore"].replace(numpy.NaN, avg_bore, inplace = True)

# Replacing missing values in "stroke"
avg_stroke = file["stroke"].astype("float").mean()
file["stroke"].replace(numpy.NaN, avg_stroke, inplace = True)

# Replacing missing values in "peak-rpm"
avg_peakrpm = file["peak-rpm"].astype("float").mean()
file["peak-rpm"].replace(numpy.NaN, avg_peakrpm, inplace = True)

# Replacing missing values in "horsepower" 
avg_hp = file["horsepower"].astype("float").mean()
file["horsepower"].replace(numpy.NaN, avg_hp, inplace = True)

# Replacing the missing values in "symboling" with mean 

avg_sym = file["symboling"].astype("float").mean()
file["symboling"].replace(numpy.NaN, avg_sym, inplace = True)
```

```python
# Checking for any more Null values in the dataset 
file.isnull().sum()
```

```python
# delete all row where price is missing 
file.dropna(subset = ["price"], inplace = True, axis = 0)

# Resetting index 
file.reset_index(drop=True, inplace=True)
```

```python
# Checking for any more Null values in the dataset 
file.isnull().sum()
```

```python
# changing the data type to float
file[["stroke", "bore", "peak-rpm"]] = file[["stroke", "bore", "peak-rpm"]].astype("float")

# changing the data type for Normalized losses to Int 
file["normalized-losses"] = file["normalized-losses"].astype("int")
```

# Performing Data Standardization 

```python
# Standardizing the fuel consumption from mpg to l/100km 
# create new column
file['city-L/100km'] = 235.215 / file['city-mpg']

# drop original column
file.drop('city-mpg', axis=1, inplace=True)
```

```python
# Standerdizing the fuel consumption on Highways from mpg to l/100km
# create new column
file['highway-L/100km'] = 235.215 / file['highway-mpg']

# drop original column
file.drop('highway-mpg', axis=1, inplace=True)
```

# Performing Data Normalization 

```python
# normalizing the width, height and length 
file["length"] = file["length"]/file["length"].max() 
file["height"] = file["height"]/file["height"].max()
file["width"] = file["width"]/file["width"].max()
file["peak-rpm"] = file["peak-rpm"]/file["peak-rpm"].max()
```

# Performing Binning 

```python
## converting data to suitable format 
file["horsepower"] = file["horsepower"].astype("float", copy=True)
```

```python
## visualizing horsepower distribution
matplt.hist(file["horsepower"])

matplt.xlabel("Horsepower of the car")
matplt.ylabel("Number of cars")
matplt.title("Horsepower distribution")
```

```python
# Creating bins 
bin = numpy.linspace(min(file["horsepower"]), max(file["horsepower"]),4)
bin
```

```python
# Assigning names to same 
names = ["low", "Medium", "High"]
file["horsepower-bin"] = pd.cut(file["horsepower"], bin, labels=names, include_lowest=True)
file[["horsepower", "horsepower-bin"]].head(5)
```

```python
# Finding the number of Vehicles in the bin 
file["horsepower-bin"].value_counts()
```

```python
# Plotting the distribution 
matplt.bar(names, file["horsepower-bin"].value_counts())
matplt.xlabel('bin')
matplt.ylabel("Number of cars")
matplt.title("Bin Vizualization")
```
